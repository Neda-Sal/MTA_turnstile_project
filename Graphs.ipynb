```python
import pandas as pd
import plotly.graph_objects as go

#grab datasets
df_left = pd.read_pickle('original_turnstile_df.pkl')
df_right = pd.read_csv('http://web.mta.info/developers/data/nyct/subway/Stations.csv')
```

```python
df_left.columns = [column.strip() for column in df_left.columns]
```

```python
right_stations = [station1.upper() for station1 in df_right['Stop Name'].unique()]
left_stations = [station2 for station2 in df_left.groupby(['STATION']).sum().reset_index()['STATION']]
```

```python
len(left_stations)
```

```python
len(right_stations)
```

```python
og_stations_df = pd.DataFrame(left_stations, columns =['Stations'])
new_stations_df = pd.DataFrame(right_stations, columns =['Stations'])
```

```python
og_stations_df
```

```python
new_stations_df
```

```python
df_og_merged = pd.merge(left = df_left, right = og_stations_df, how="right", left_on="STATION", right_on="Stations")
df_new_merged = pd.merge(left = og_stations_df, right = new_stations_df, how="inner", left_on="Stations", right_on="Stations")
```
